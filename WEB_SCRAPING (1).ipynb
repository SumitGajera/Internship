6) Please visit https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloadedarticles/ and scrap
a) Paper title
b) date
c) Author

```python
r = requests.get('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')


print(r)


# print(r.content)

soup = BeautifulSoup(r.content, 'html.parser')
# print(soup.prettify())

# links = soup.findAll('tr', class_='athing')
posts = soup.find_all('div', class_='post')
print(posts)


title = soup.find('h2', class_="h5 article-title")
title
title.text

date = soup.find('p' , class_ ="article-date")
date
date.text

author = soup.find('p', class_="article-authors")
author
author.text
```

```python
titles =[]
for i in soup.find_all('h2', class_="article-title"):
  titles.append(i.text)

titles
```

```python
dates =[]
for i in soup.find_all('p' , class_ ="article-date"):
  dates.append(i.text)

dates
```

```python
authors =[]
for i in soup.find_all('p', class_="article-authors"):
  authors.append(i.text)

authors
```

```python
print(len(titles) ,len(dates) , len(authors))
a = {'Title' :titles ,'Date' : dates ,'Author' : authors}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
# df = pd.DataFrame({'Titles' :titles ,'Location' : location ,'Price' : price})
df
```
